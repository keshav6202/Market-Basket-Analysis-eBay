# Task 3 :- Customer Segmentation and Profiling

### Segment customers based on purchase frequency and shopping satisfaction levels

```python
def segment_customer(row):

    high_frequency = [
        "Multiple times a day",
        "Multiple times a week"
    ]

    medium_frequency = [
        "Once a week",
        "Few times a month"
    ]

    if (
        row["Purchase_Frequency"] in high_frequency
        and row["Shopping_Satisfaction"] >= 4
    ):
        return "Frequent Buyer"

    elif (
        row["Purchase_Frequency"] in medium_frequency
        and row["Shopping_Satisfaction"] >= 3
    ):
        return "Occasional Shopper"

    else:
        return "At-Risk Customer"

df["Customer_Segment"] = df.apply(
    segment_customer,
    axis=1
)
```

```python
segment_count = df["Customer_Segment"].value_counts()

print(segment_count)
```

```python
import matplotlib.pyplot as plt

segment_count.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Customer Segments")
plt.xlabel("Segment")
plt.ylabel("Number of Customers")

plt.show()
```

# K-Means Clustering

```python
features = df[
    [
        "Customer_Reviews_Importance",
        "Rating_Accuracy",
        "Shopping_Satisfaction"
    ]
]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_features = scaler.fit_transform(
    features
)
```

```python
# Apply K-Means

from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=3,
    random_state=42
)

df["Cluster"] = kmeans.fit_predict(
    scaled_features
)
```

```python
print(
    df["Cluster"].value_counts()
)
```

```python
df["Cluster"].value_counts().plot(
    kind="bar"
)

plt.title("Cluster Distribution")

plt.show()
```

```python
cluster_profile = df.groupby(
    "Cluster"
)[
    [
        "Shopping_Satisfaction",
        "Rating_Accuracy",
        "Customer_Reviews_Importance"
    ]
].mean()

print(cluster_profile)
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))

sns.scatterplot(
    x="Rating_Accuracy",
    y="Shopping_Satisfaction",
    hue="Cluster",
    data=df,
    palette="Set1"
)

plt.title("Customer Clusters")

plt.show()
```
